```python
d_model=6
max_sequence_len=11
```

as the embeddings, so that the two can be summed. There are many choices of positional encodings,

```python
pos=torch.arange(max_sequence_len,dtype=torch.float) #for sentence we have a sentence of 11len
pos
```

```python
pos.shape
```

```python
pos=pos.reshape(max_sequence_len,1) #can use unsqueeze too
```

```python
pos
```

```python
pos.shape
```

```python
even_i= torch.arange(0,d_model,2) #0 to dmodel with 2 steps
```

```python
even_i.shape
```

```python
even_denominator= torch.pow(10000,even_i/d_model)
```

```python
even_denominator
```

```python
odd_i=torch.arange(1,d_model,2)
odd_i
```

```python
odd_denominator=torch.pow(10000,(odd_i-1)/d_model)
odd_denominator
```

odd_denominator === even_denominator, so we can use either of them


```python
denominator=even_denominator
denominator
```

```python
even_PE= torch.sin(pos/denominator)
even_PE
```

```python
even_PE.shape
```

```python
odd_PE=torch.cos(pos/denominator)
```

```python
odd_PE
```

```python
odd_PE.shape
```

```python
stacked= torch.stack([even_PE,odd_PE], dim=2)
stacked
```

```python
stacked.shape
```

```python
PE= torch.flatten(stacked,start_dim=1,end_dim=2)
```
